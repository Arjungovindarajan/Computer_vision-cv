# Arithmetic Operation with image

```python
imgs_1 = cv2.resize(img_dummy_1.astype(float), (780, 540))
imgs_2 = cv2.resize(img_dummy_2.astype(float), (780, 540))
```

```python
# result = img_1 + img_2
# plt.imshow(result)
result_add = imgs_1 + imgs_2
plt.imshow(result_add)
```

```python
result_sub = imgs_1 - imgs_2
plt.imshow(result_sub)
```

```python
result_multi = imgs_1 * imgs_2
plt.imshow(result_multi)
```

```python
result_div = imgs_1 / imgs_2
plt.imshow(result_div)
```

```python
image = cv2.convertScaleAbs(imgs_1, alpha=1.0/255.0)
```

```python
figure_size = 7
blur_image = cv2.blur(image, (figure_size, figure_size))
plt.figure(figsize=(11, 6))
plt.subplot(121), plt.imshow(image), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur_image), plt.title("Blurred")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
new_img = cv2.imread("/content/new_obj.JPG")
plt.imshow(new_img)
```

```python
(height, width) = new_img.shape[:2]
zoom = cv2.resize(new_img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
# zoom image
zoom = cv2.resize(new_img, dsize=(int(width*2), int(height*2)), interpolation=cv2.INTER_LINEAR)
plt.imshow(zoom)
```

#Rotation image

```python
rot_img = cv2.getRotationMatrix2D((width/2, height/2),30,1)
rotate = cv2.warpAffine(new_img, rot_img, (height, width))
plt.imshow(rotate)
```

```python
r = np.float32([[1,0,50],[0,1,100]])
dst = cv2.warpAffine(new_img, r,(height, width))
plt.imshow(dst)
```
